```python
df=pd.read_csv('../input/graduate-admissions/Admission_Predict.csv')
```

```python
df.head(5)
```

```python
df.isnull().sum()        ###To check Null values
```

```python
x=df.iloc[:,1:8].values
y=df.iloc[:,8].values
```

```python
##standardize

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=3)
```

```python
print(y_train.shape)
x_train.shape
```

```python
import tensorflow as tf
tf.keras.__version__
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(3, input_dim=x_train.shape[1], kernel_initializer='normal', activation='relu',name='hidden_layer'))
#model.add(tf.keras.layers.Dense(3, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='linear'))
model.summary()
```

```python
model.compile(loss='mse',optimizer='adam',metrics=['mse','mae'])
```

```python
model.fit(x_train,y_train,epochs=300,batch_size=10,verbose=1)
```

```python
y_pred=model.predict(x_test)        ###changing epochs,increasing node,increasing hidden layer will change score
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
